```python
# importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#loading the dataset and and reading
df = pd.read_csv('housing_data.csv')
df.head()
```

```python
#checking the null value
df.isnull().sum()
```

```python
#checking all the columns
df.columns
```

```python
#calculating shape of the data
df.shape
```

```python
# getting some information about the data
df.info()
```

```python
#Because we have already index column
df.drop(['Unnamed: 0'],axis=1 , inplace = True)
```

```python
df.info()
```

```python
#Dropping MasVnrType and Alley beacuse most of the data are missed
df.drop(['Alley' , 'MasVnrType'], axis = 1 , inplace = True)
df.info()
```

```python
df['Exterior1st'].unique()
```

```python
df['Exterior2nd'].unique()
```

```python
# Dropping Exterior2nd because both are containing same information
df.drop(['Exterior2nd'], axis =1 ,  inplace = True)
```

```python
#checking null values
df['GarageYrBlt'].isnull().sum()
```

```python
#filling null value with mean
df['GarageYrBlt'] = df['GarageYrBlt'].fillna(df['GarageYrBlt'].mean())
```

```python
df['GarageYrBlt'].isnull().sum()
```

```python
df.info()
```

```python
#checkinh null value
df['Electrical'].isnull().sum()
```

```python
# filling null value with mode
df['Electrical'] = df['Electrical'].fillna(df['Electrical'].mode()[0])
```

```python
df['Electrical'].isnull().sum()
```

```python
df.info()
```
